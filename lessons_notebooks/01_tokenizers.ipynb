```python

import re
import pandas as pd
from collections import Counter
import spacy
import nltk
from transformers import AutoTokenizer

path_en_core_web_sm = 'C:/Users/aseva/AppData/Local/uv/cache/archive-v0/tAkJPRrotU1RpBeNdm3Yz/en_core_web_sm/en_core_web_sm-3.8.0'
```

```python
with open("../data/tiny_shakespear.txt", "r") as f:
    text = f.read()

# регулярное выражение - оставляет только слова длиной  > 1 символа без пунктуации
pattern = r"(?u)\b\w\w+\b" 

re.findall(pattern, text[:60]) 
```

```python

# для некоторых библиотек требуется ручной вызов загрузки необходимых словарей и модулей
#nltk.download('punkt')
#nltk.download('punkt_tab')

#в nltk указываем язык, с которым работаем явно
tokens_nltk = nltk.word_tokenize(text[:60], language='english') 
```

```python

# для spacy загрузим отдельную модель/словарь для английского языка en_core_web_sm    
nlp = spacy.load(path_en_core_web_sm)

# для spacy предварительно оборачиваем текст во внутренний формат
# представления документа и используем атрибут .text у токена
doc = nlp(text[:60])
tokens_spacy = [token.text for token in doc]
```

```python

# в transformers используем предобученный токенизатор от модели для английского языка
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

# токенайзеры transfomers работают с текстами напрямую
tokens_trf = tokenizer.tokenize(text[:60])
```

```python

print("NLTK токены        :", tokens_nltk)
print("transformers токены:", tokens_trf )
print("spaCy токены       :", tokens_spacy)
```

# Задание 1

```python
text = "123: hey ds_expert@w.com , check findme.com"

# загрузите модель для английского корпуса из spacy, models/en_core_web_sm
nlp = spacy.load(path_en_core_web_sm)

# преобразуйте текст во внутренний формат spacy
doc = nlp(text)
tokens = [token.text for token in doc]

# проверьте, является ли токен числом, имейлом, ссылкой
digits = [token.is_digit for token in doc]
emails = [token.like_email for token in doc]
urls = [token.like_url for token in doc]
```

```python
pd.DataFrame(
    columns=tokens,
    index=['is_digits', 'is_emails', 'is_urls'],
    data=[digits, emails, urls]
).map(int)
```

# Stop Words

```python
with open("../data/tiny_shakespear.txt", "r") as f:
    text = f.read()

tokens_regex = re.findall(pattern, text.lower())

cnt = Counter(tokens_regex)
print(cnt.most_common(5))
print(cnt.most_common()[-5:]) 
```

```python
#nltk.download('stopwords') # подгружаем стоп-слова

from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english')) # выбираем нужный язык
tokens_filtered = [t for t in tokens_regex if t not in stop_words] # фильтруем

cnt = Counter(tokens_filtered)
print(cnt.most_common(5))
print(cnt.most_common()[-5:]) 
```

```python
from gensim.models import FastText
```

```python

fasttext_path = '../models/cc.ru.300.bin.gz'                 # 300 - размер вектора слова
model = FastText.load_fasttext_format(fasttext_path)
```

```python
result = model.wv.most_similar("ученый", topn=5) 

for word, sim in result:
    print(word.ljust(20), round(sim,5))
```

# еще примеры

### найдём похожие слова для OOV слова «лаборантишка» через FastText:
Разбивая исходное слово на n-gram’ы и вычисляя средний вектор слова, FastText подобрал верных соседей — смысл слов корректно закодирован в этих векторах.

```python
result = model.wv.most_similar("лаборантишка", topn=5) 

for word, sim in result:
    print(word.ljust(20), round(sim,5))
```

### Посмотрим на пример работы с опечатками, имитируем случайное залипание клавиши «н»

```python
result = model.wv.most_similar("ученный", topn=5) 

for word, sim in result:
    print(word.ljust(20), round(sim,5))
```

но не все коту масленница - иногда и не работает:

```python
result = model.wv.most_similar("учееный", topn=5) 

for word, sim in result:
    print(word.ljust(25), round(sim,5))
```

### В реализации `FastText` в `gensim` также есть возможности, позволяющие оценить согласованность векторов слов. Делается это с помощью метода  `wv.doesnt_match()`:

```python
text = "data science это good"
model.wv.doesnt_match(text.split(" ")) 
```

```python
import numpy as np
from gensim.models import KeyedVectors
from scipy.spatial.distance import cosine
```

```python
# укажем путь к предзагруженной матрице
word2vec_path = '../models/word2vec-google-news-300.gz'

# загружаем веса
w2v = KeyedVectors.load_word2vec_format(word2vec_path, binary=True) 
```

```python
for word, sim in w2v.most_similar("mother", topn=5):
    print(word.ljust(12), round(sim,5))
```

```python
result = w2v.most_similar(
    positive=['fairy', 'woman'],
    negative=['magic'],
    topn=5
)
for word, sim in result:
    print(word.ljust(15), round(sim,5))
```

```python
result = w2v.most_similar(
    positive=['king', 'woman'],
    negative=['man'],
    topn=5
)
for word, sim in result:
    print(word.ljust(15), round(sim,5))
```

* Сравните полученную косинусную близость двух пар предложений (код для этого уже написан).
* Используйте метод `get_vector()` для получения эмбеддинга слова.

```python


def get_sentence_embedding(tokens, w2v_model):
    vectors = []
    for t in tokens:
        vectors.append(w2v_model.get_vector(t))
    return np.mean(vectors, axis=0)

  
tokens_A = ["data", "scientists", "are", "cool"]
tokens_B = ["data", "science", "is", "awesome"]
tokens_C = ["clouds", "in", "the", "sky"]

emb_A = get_sentence_embedding(tokens_A, w2v)
emb_B = get_sentence_embedding(tokens_B, w2v)
emb_C = get_sentence_embedding(tokens_C, w2v)

float(1 - cosine(emb_A, emb_B)), float(1 - cosine(emb_A, emb_C))
```

```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
```

```python

with open("../data/tiny_shakespear.txt", "r") as f:
    text = f.read()

# разобьём текст на части - шаг необязательный, 
# просто искусственно делаем из 1 документа 10 разных)
step = 100000
docs = [text[i:i+step] for i in range(0, len(text), step)]
```

```python
stop_words = stopwords.words('english')

tf_idf = TfidfVectorizer(stop_words=stop_words) # сразу передадим стоп-слова

# используем комбинацию методов fit и transform
tf_idf_matrix = tf_idf.fit_transform(docs)
```

```python
print("N_documents x Vocabulary_size:", tf_idf_matrix.shape, '\n')
pd.DataFrame(
    index=['token value','token weight'],
    data=[
          list(pd.Series(tf_idf.get_feature_names_out()[:5]).map(lambda s: s.ljust(7))),   # 5 токенов из словаря
          list(pd.DataFrame(tf_idf_matrix[0, :5].todense()).iloc[0])                       # веса этих 5 токенов
    ]
).T
```

Постройте матрицу TF-IDF с такими особенностями:
* Ограничьте максимальный размер словаря 10000.
* Используйте стоп-слова и одно+двухсловесные словесные ngram’ы одновременно.
* Ограничьте минимальную частотность слов двумя словами, а долю самых частых слов сверху в 95%.
* Добавьте стоп-слова, а также ограничения из условия в параметры TfidfVectorizer: `ngram_range=(1, 2)`, `min_df=2`, `max_df=0.95`, `max_features=10000`

```python
step = 100000
docs = [text[i:i+step] for i in range(0, len(text), step)]

tf_idf = TfidfVectorizer(analyzer="word",
                         stop_words=stop_words,
                         ngram_range=(1, 2),      # 1-2 словесные n-gramы
                         min_df=2,                # токен не реже, чем в 2 документах
                         max_df=0.95,             # не учитываем токен с встречаемостью > 95%
                         max_features=10000       # размер словаря
)

tf_idf_matrix = tf_idf.fit_transform(docs)
```

```python
print("N_documents x Vocabular_size:", tf_idf_matrix.shape)
pd.DataFrame(
    index=['token value','token weight'],
    data=[
          list(pd.Series(tf_idf.get_feature_names_out()[:5]).map(lambda s: s.ljust(7))),   # 5 токенов из словаря
          list(pd.DataFrame(tf_idf_matrix[0, :5].todense()).iloc[0])                       # веса этих 5 токенов
    ]
).T
```
